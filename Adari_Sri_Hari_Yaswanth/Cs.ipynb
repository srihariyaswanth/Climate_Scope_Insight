Data Analysis


```python
#Descriptive Statistical Analysis
statistical_summary=df.describe()
statistical_summary
```

```python
#Analyzing Temperature by Country
country_groups = df.groupby('country')
avg_temp_by_country = country_groups['temperature_celsius'].mean()
avg_temp_by_country
```

```python
# Sort the results from highest to lowest to see the warmest countries.
warmest_countries = avg_temp_by_country.sort_values(ascending=False)
print("--- Top 10 Warmest Countries (Average Temperature) ---")
print(warmest_countries.head(10))
```

```python
# Sort the results from lowest to highest to see the coldest countries.
coldest_countries = avg_temp_by_country.sort_values()
print("\n--- Top 10 Coldest Countries (Average Temperature) ---")
print(coldest_countries.head(10))
```

```python
numeric_cols=df.select_dtypes(include=['float64','int64'])
correlation_matrix=numeric_cols.corr()
print("Correlation matrix : ")
correlation_matrix
```

```python
# This turns the grid-like matrix into a simple list
corr_unstacked = correlation_matrix.unstack()

# Convert the list to a DataFrame and name the columns
corr_pairs = corr_unstacked.reset_index()
corr_pairs.columns = ['variable_1', 'variable_2', 'correlation']

# Remove the uninteresting correlations of a variable with itself (which are always 1)
corr_pairs = corr_pairs[corr_pairs['variable_1'] != corr_pairs['variable_2']]

# Remove duplicate pairs
corr_pairs['pair_key'] = corr_pairs.apply(lambda row: tuple(sorted((row['variable_1'], row['variable_2']))), axis=1)
corr_pairs = corr_pairs.drop_duplicates(subset=['pair_key']).drop(columns=['pair_key'])
strongest_positive = corr_pairs.sort_values(by='correlation', ascending=False)
print("--- Top 5 Strongest Positive Correlations ---")
print(strongest_positive.head(5))
strongest_negative = corr_pairs.sort_values(by='correlation', ascending=True)
print("\n--- Top 5 Strongest Negative Correlations ---")
print(strongest_negative.head(5))
```
